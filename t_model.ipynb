```python
import pandas as pd
df=pd.read_csv("Reviews.csv")
df.head()
```

```python
Id=[]
ProductId=[]
UserId=[]
ProfileName=[]
HelpfulnessNumerator=[]
Score=[]
Time=[]
Summary=[]
Text=[]
count=0
for i in range(0,40000):
    if df['Score'][i] < 3:
        #Id.append(df['Id'][i])
        ProductId.append(df['ProductId'][i])
        UserId.append(df['UserId'][i])
        ProfileName.append(df['ProfileName'][i])
        HelpfulnessNumerator.append(df['HelpfulnessNumerator'][i])
        Score.append('0')
        Time.append(df['Time'][i])
        Summary.append(df['Summary'][i])
        Text.append(df['Text'][i].lower())
        count+=1
        if count==5000:
            break
for i in range(0,40000):
        if df['Score'][i] >= 3:
        #Id.append(df['Id'][i])
            ProductId.append(df['ProductId'][i])
            UserId.append(df['UserId'][i])
            ProfileName.append(df['ProfileName'][i])
            HelpfulnessNumerator.append(df['HelpfulnessNumerator'][i])
            Score.append('1')
            Time.append(df['Time'][i])
            Summary.append(df['Summary'][i])
            Text.append(df['Text'][i].lower())
            count+=1
            if count==10000:
                break
```

```python
df = pd.DataFrame({'ProductId': ProductId, 'UserId': UserId, 'ProfileName': ProfileName, 'HelpfulnessNumerator': HelpfulnessNumerator, 'Score':Score, 'Time':Time, 'Summary':Summary,'Text':Text })
df.to_csv('final.csv',index=False)
```
